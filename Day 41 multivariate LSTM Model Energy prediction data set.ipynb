```python
# Create data frame with selected columns

df_input=df[['Appliances','lights','T_out','RH_1','Visibility']]
display(df_input)
```

```python
# MIn Max Scaler

'''Since thre are outerliers in the data use min max scaler to scale the data.
   Scaler helps to coverage the gradients faster.'''
   
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data_scaled=scaler.fit_transform(df_input)
display(data_scaled)   
```

```python
# Independant variable

features=data_scaled
display(features)
```

```python
features.shape
```

```python
# Dependant variable

target=data_scaled[:,0]
display(target)
```

```python
# Display top 20 records

pd.DataFrame(features).head(20)
```

```python
# Time Series Generator
'''Use two previous periods data to get the predicted value
   Features: Independent Variable
   Target: Dependent Variable
   Length=2 (Number of previous periods used to predict)
   Sampling rate: Number of records from sample 1 means all records, 2 means alternate records
   Batch Size: Number of sets 1 Means one set, Two means two sets
'''

from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
TimeseriesGenerator(features,target,length=2,sampling_rate=1,batch_size=1)[0]
```

```python
# Train Test split without shuffle

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=.2,random_state=123,shuffle=False)
display(x_train.shape)
display(x_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
# Time Series Generator - Train

WS=720 # use 5 days of data prediction 6*24*5=720
batch_size=32 # Pass 32 data set at a time
num_features=4
train_generator=TimeseriesGenerator(x_train,y_train,length=WS,sampling_rate=1,batch_size=batch_size)
display(train_generator[0]) # This display one batch all the required input for 32 prediction
```

```python
# Time Serious Generator - Test

test_generator=TimeseriesGenerator(x_test,y_test,length=WS,sampling_rate=1,batch_size=batch_size)
display(test_generator[0])
```

```python
# Import Libraries

from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import LeakyReLU
from keras.layers import Dense
from keras.layers import Dropout 
```

```python
# Create sequential model input Layer

# Create sequential model

model=Sequential()

# LSTM Layer - Input Layer

# input_shape 720*4 features ie 720 records ie 5 days data to do prediction

model.add(LSTM(128,input_shape=(WS,num_features),return_sequences=True))

# LeakyReLU Activation function

model.add(LeakyReLU(alpha=0.5))
```

```python
# Middle LSTM Layer

model.add(LSTM(128,return_sequences=True))
model.add(LeakyReLU(alpha=0.5))
model.add(Dropout(0.3)) # Drop out to avoid overfitting
```

```python
# LSTM Layer with 64 neurons

# return_sequences= False since this is the last layer

model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.3))
```

```python
# Output Layer

model.add(Dense(1))
```

```python
# Display Summary

model.summary()
```

```python
# Define Early Stopping

# if the val_loss isnot improving in two iteration stop the execution

from keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(monitor='val_loss',patience=2,mode='min')
```

```python
from tensorflow.keras.metrics import MeanAbsoluteError
```

```python
# Compile the Model

# Adam optimizer good for time series

model.compile(optimizer='adam',loss='mean_squared_error',metrics=[MeanAbsoluteError()])
```

```python
# Model Fit 

history=model.fit_generator(train_generator,epochs=50,validation_data=test_generator,shuffle=False,callbacks=[early_stopping])
```
